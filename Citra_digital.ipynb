```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# upload gambar
from google.colab import files
uploaded = files.upload()

# baca gambar grayscale
img = cv2.imread(list(uploaded.keys())[0], cv2.IMREAD_GRAYSCALE)

# fungsi kuantisasi
def quantize(image, levels):
    step = 256 // levels
    quantized = (image // step) * step
    return quantized

# kuantisasi
img_2bit = quantize(img, 4)      # 4 level
img_4bit = quantize(img, 16)     # 16 level
img_8bit = quantize(img, 256)    # 256 level

# tampilkan
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("Original")
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title("2-bit (4 level)")
plt.imshow(img_2bit, cmap='gray')

plt.subplot(2,2,3)
plt.title("4-bit (16 level)")
plt.imshow(img_4bit, cmap='gray')

plt.subplot(2,2,4)
plt.title("8-bit (256 level)")
plt.imshow(img_8bit, cmap='gray')

plt.show()
```
