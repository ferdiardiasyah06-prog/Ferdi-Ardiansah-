```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread(list(uploaded.keys())[0], 0)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()
```

```python
negatif = 255 - img
```

```python
plt.imshow(negatif, cmap='gray')
plt.title('Negatif')
plt.axis('off')
plt.show()
```

```python
c = 255 / np.log(1 + np.max(img))
log_img = c * np.log(1 + img)
```

```python
log_img = np.log1p(img.astype(np.float64))

# NORMALISASI ke 0–255 (pakai min-max biar stabil)
log_img = (log_img - np.min(log_img)) / (np.max(log_img) - np.min(log_img)) * 255

log_img = log_img.astype(np.uint8)
```

```python
plt.imshow(log_img, cmap='gray')
plt.title('Log Transform FINAL')
plt.axis('off')
plt.show()
```

```python
print("Original max:", np.max(img))
print("Original min:", np.min(img))

print("Log max:", np.max(log_img))
print("Log min:", np.min(log_img))
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Cita Original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(negatif, cmap='gray')
plt.title('Cita Negatif')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(log_img, cmap='gray')
plt.title('Cita Logaritmik')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(img.ravel(), bins=256)
plt.title('Histogram Original')

plt.subplot(1,3,2)
plt.hist(negatif.ravel(), bins=256)
plt.title('Histogram Negatif')

plt.subplot(1,3,3)
plt.hist(log_img.ravel(), bins=256)
plt.title('Histogram Logaritmik')

plt.show()
```
